```python
from queue import Queue
def stutter(x: Queue,num):
    l = []
    for j in range(x.qsize()):
        a = x.get()
        for i in range(num):
            x.put(a)
    while(x.empty() == False):
        l.append(x.get())
    print(l)
q = Queue()

while(True):
    try:
        inp = int(input("Enter the Elemeents for Queue:(string to exit input): "))
    except:
        break
    if(inp<1):
        print("Not possible")
        q.clear()
        break
    else:
        q.put(inp)
num = int(input("Enter the number of number of times you need to print the values: "))   
stutter(q,num)
```
