```python
// Java program for the above approach
import java.util.*;
// intializing the main function
class Main{
    // method to check the right angle is forming or not
    static void rightAngledCheck(int X1, int Y1,
                                 int X2, int Y2,
                                 int X3, int Y3){
         
        // calculating the sides 
        int A = (int)Math.pow((X2 - X1), 2) +
                (int)Math.pow((Y2 - Y1), 2);
     
        int B = (int)Math.pow((X3 - X2), 2) +
                (int)Math.pow((Y3 - Y2), 2);
     
        int C = (int)Math.pow((X3 - X1), 2) +
                (int)Math.pow((Y3 - Y1), 2);
     
        // checking Pythagoras Formula
        if ((A > 0 && B > 0 && C > 0) &&
            (A == (B + C) || B == (A + C) ||
             C == (A + B)))
             // displaying True if the traingle is formed
            System.out.println("True");
        else
        // displaying False if the traingle is formed
            System.out.println("False");
    }
 
    // intializing the main function
    public static void main(String s[]){
        // local varibles
        int X1, Y1, X2, Y2, X3, Y3;
        // intializing the Scanner class
        Scanner sc = new Scanner(System.in);
        // getting values from user
        X1 = sc.nextInt();
        Y1 = sc.nextInt();
        X2 = sc.nextInt();
        Y2 = sc.nextInt();
        X3 = sc.nextInt();
        Y3 = sc.nextInt();
        // function calling
        rightAngledCheck(X1, Y1, X2, Y2, X3, Y3);
    }
}
```
